```python
from sklearn.utils import resample
df_majority = df[df['Class'] == 1]
df_minority = df[df['Class'] == 0]
```

```python
df_minority_upsampled = resample(df_minority,
                                replace = True,
                                n_samples = 250,
                                random_state = 123)
```

```python
df_final = pd.concat([df_majority, df_minority_upsampled])
```

```python
sns.countplot(data = df_final, x = 'Class')
```

```python
import scipy.stats as stats
```

```python
z = np.abs(stats.zscore(df_final))
```

```python
z.describe()
```

```python
df_clean = df_final[(z < 3).all(axis = 1)]
df_clean.shape
```

```python
sns.heatmap(df_clean.corr())
```

```python
X = df_clean.drop('Class', axis = 1)
y = df_clean['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_rf_predict = rf.predict(X_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_rf_predict))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_rf_predict))
print('Classification Report:\n', classification_report(y_test, y_rf_predict))
```
